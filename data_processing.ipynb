# Handing for Date

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year, month, day, and hour
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
```

```python
conditions = [
    (df['Hour'] >= 7) & (df['Hour'] < 12),
    (df['Hour'] > 12) & (df['Hour'] < 14),
    (df['Hour'] >= 14) & (df['Hour'] < 18),
    (df['Hour'] >= 18) & (df['Hour'] < 21),
    (df['Hour'] < 7) & (df['Hour'] >= 21)
]

choices = ['Morning', 'Noon', 'Afternoon', 'Evening', 'Night']

df['Time_of_Day'] = np.select(conditions, choices, default='Night')
```

```python
df['week'] = df['Date'].dt.day_name()
```
